## Explanations
### Random state

Random state number is used for initializing the internal random number generator, which will decide the splitting of data into train and test indices. It should be set the same value, in order to validate result processing over multiple runs of the code.

Random state numbers of 0, 1, 7, 10, 15, 35 & 42 were tested and it was observed that these different random state numbers do not give the same results of evaluation metric performance, **random state number of 7** gave the highest overall accuracy and F1_Score.


### It should be noted that the Precision, Recall and F1 scores are weighted averages of the scores for the several vehicles (opel, bus, saab, van).

**Focusing on the results with random state number being equal to 7**

The scores show that a train/test split ratio of 0.9/0.1 respectively has given the highest accuracy score of 83.5% and F1 score of 82.8%. 

 - Precision refers to the weighted percentage of predictions that are correct.
- 0.9/0.1 train/test ratio with Precision score 85.52%, means 85.52% of the predictions are correct.
    
    
 - Recall refers to the weighted percentage(of all vehicles) of each of the postive cases caught.
- 0.9/0.1 train/test ratio with Recall score, 83.53%, means  83.53% of postive cases are caught in total(that is for all the vehicles).


 - F1 Score refers to the weighted percentage of scores in which the positive predictions made were correct.
- So, for a 0.9/0.1 train/test ratio, 82.82% of the positive predictions made are correct.

```python
# creating variables for some of the columns in the dataframe using the results_7 (random state number being equal to 7).
test_size = results_7["Test Size"]
accuracy = results_7["Accuracy Score(%)"]
f1 = results_7["F1_Score(%)"]
```

```python
# Graph of accuracy score against test size
fig = plt.figure(figsize=(14, 7))
plt.plot(test_size, accuracy)
fig.suptitle(
    "Graph showing the relationship between the test sizes and accuracy of the models",
    fontsize=20,
)
plt.xlabel("Test sizes", fontsize=16)
plt.ylabel("Accuracy(%)", fontsize=16)
plt.grid(axis="x")
plt.show()
```

```python
# Graph of F1 score against test size
fig = plt.figure(figsize=(14, 7))
plt.plot(test_size, f1, color="red")
fig.suptitle(
    "Graph showing the relationship between the test sizes and F1_score of the models",
    fontsize=20,
)
plt.xlabel("Test sizes", fontsize=16)
plt.ylabel("F1_Score(%)", fontsize=16)
plt.grid(axis="x")
plt.show()
```

```python
# Graph of Accuracy and F1_score against test size
fig = plt.figure(figsize=(14, 7))
plt.plot(test_size, accuracy, test_size, f1)
plt.legend(["Accuracy", "F1_Score"])
fig.suptitle(
    "Graph showing the relationship between the test sizes, accuracy and F1 score of the models",
    fontsize=20,
)
plt.xlabel("Test sizes", fontsize=16)
plt.ylabel("Scores in %", fontsize=16)
plt.ylim(72, 85)
plt.grid(axis="x")
plt.show()
```

### Conclusion
I have been able to show how the metric performance depends on our choice of train/test split ratio and I have also learnt that the value of the random state used in splitting the test size has an effect on the performance of the model.
- Looking at the graph, the accuracy scores are directly proportional to F1_scores using the results with the random state number = 7. 
- It is also observed that these scores reduce as the test_size increase except a slight increase in performance with test sizes 0.5 and 0.6. The highest score was at a train/test split ratio of 0.9/0.1 followed by 0.4/0.6, then 0.7/0.3 accordingly.

Therefore, I'll conclude that 0.9/0.1 or 0.7/0.3 are good train/test split ratios and yield a better model performance.
The general performance of this Logistic Regression model on this Vehicles dataset improves with a reduced test size and declines with an increased test size (especially with using test_sizes between 0.6 and 0.9).
